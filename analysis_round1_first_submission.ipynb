```python
import pandas as pd

# Load the file again but skip the first row if it's not the header
df = pd.read_csv('log_data_r1.csv', sep=';', header=1)

df
```

```python
# Clean column names just in case
df.columns = df.columns.str.strip()

# Convert profit column to numeric
df['profit_and_loss'] = pd.to_numeric(df['profit_and_loss'], errors='coerce')

# Group and sum PnL by product
pnl_by_product = df.groupby('product')['profit_and_loss'].sum().sort_values(ascending=False)

# Show the result
print(pnl_by_product)
```
